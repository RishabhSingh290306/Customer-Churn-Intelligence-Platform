```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df["TotalCharges"].dtype
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df["TotalCharges"].isnull().sum()
```

## TotalCharges Cleaning

The TotalCharges column was incorrectly stored as a string datatype.

Investigation revealed that 11 records contained blank values. All 11 customers had a tenure of 0 months, indicating they were newly acquired customers who had not yet accumulated any charges.

The column was converted to numeric format and missing values were replaced with 0, preserving customer records while maintaining business logic.

# Data Cleaning Summary

1. No missing values were found in the dataset.

2. No duplicate records were identified.

3. The TotalCharges column was incorrectly stored as a string datatype.

4. Investigation revealed 11 blank values belonging to customers with tenure = 0 months.

5. TotalCharges was converted to numeric format and missing values were replaced with 0.

6. The cleaned dataset was exported for further analysis.
